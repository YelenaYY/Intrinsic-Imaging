```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# plot the validation loss of decomposer

train_1 = pd.read_csv("logs/train_decomposer/validate_2025-10-13_22-48.csv")
train_2 = pd.read_csv("logs/train_decomposer/validate_2025-10-14_13-37.csv")
train_3 = pd.read_csv("logs/train_decomposer/validate_2025-10-14_21-04.csv")

# merge the three dataframes
train = pd.concat([train_1, train_2, train_3])

# plot the validation loss
plt.figure(figsize=(10, 5))
plt.plot(train["epoch"], train["validate_loss"], label="validate_loss")
plt.plot(train["epoch"], train["reflectance_loss"], label="reflectance_loss")
plt.plot(train["epoch"], train["depth_loss"], label="depth_loss")
plt.plot(train["epoch"], train["normals_loss"], label="normals_loss")
plt.plot(train["epoch"], train["lights_loss"], label="lights_loss")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.legend()
plt.title("Validation Loss of Decomposer")
plt.show()
```

```python
# shader loss
train = pd.read_csv("logs/train_shader_variant/train_shader_2025-10-21_18-24.csv")


train_variant_1 = pd.read_csv("logs/train_shader/train_shader_2025-10-21_18-07.csv").iloc[:30]

# plot the training loss
plt.figure(figsize=(5, 5))
plt.plot(train["epoch"], train["val_loss"], label="validation loss of variant")
plt.plot(train_variant_1["epoch"], train_variant_1["val_loss"], label="validation loss of original")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.legend()
```
